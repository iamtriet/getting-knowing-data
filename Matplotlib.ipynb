## Pyplot API vs Object-Oriented (OO) API
* Pyplot API -> Quickly
* OO API -> Advanced

### Pyplot API

```python
x = np.linspace(0,10,1000)
plt.plot(x, np.sin(x), color="blue", linestyle="dashed", label="sin(x)");
plt.plot(x, np.cos(x), color="red", linestyle="dashed", label="cos(x)");

plt.title("sin(x) and cos(x)")
plt.xlabel("Bien x")
plt.ylabel("Sin(x)")
#plt.xlim([0,4]) #limit x axis
#plt.ylim([-0.75,0.75]) #limit y axis
#plt.axis([0, 4, -0.75, 0.75]) #plt.axis([xmin, xmax, ymin, ymax])
#plt.axis("tight")
#plt.axis("equal")

plt.legend(); #C tc dng hin label
```

### Object-Oriented API

```python
#1. Prepare data
x = [1,2,3,4]
y = [11,22,33,44]

#2. Setup plot
fig, ax = plt.subplots(figsize=(5,5)) #Figure size = Width & Height of the plot, fig = figure object (hnh tng); ax = axes object (hnh con)

#3. Plot the data
ax.plot(x,y);

ax.set(title="A simple plot",
       xlabel="x-axis",
       ylabel="y-axis");
```

### 3.Bar

```python
#Prepare data
soft_drink_prices = {"Coke": 10,
                     "Pepsi": 12,
                     "Sprite": 8}

fig, ax = plt.subplots()
ax.bar(soft_drink_prices.keys(), soft_drink_prices.values())
ax.set(title="Bach Hoa Xanh's Soft Drink Prices", ylabel="Price ($)")
```

### 4. Histogram

```python
#Prepare data

np.random.seed(42)
student_height = np.random.normal(170, 10, 250)
```

```python
student_height[:10], len(student_height)
```

```python
# Pyplot API
plt.hist(student_height);
```

### Subplots

```python
#Option 1:
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                       ncols=2,
                       figsize=(10,5))

#Plot data to each subplot
ax1.plot(x,x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(soft_drink_prices.keys(), soft_drink_prices.values())
ax4.hist(student_height, bins=30);
```

```python
#Option 2:
fig, ax = plt.subplots(nrows=2,
                       ncols=2,
                      figsize=(10,5))

#Plot data to each subplot
ax[0,0].plot(x,x/2);
ax[0,1].scatter(np.random.random(10), np.random.random(10));
ax[1,0].bar(soft_drink_prices.keys(), soft_drink_prices.values())
ax[1,1].hist(student_height, bins=30);
```

# Project: Data Visualization - California Cities

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
cities = pd.read_csv(r'C:\Users\ASUS\OneDrive - somewhereicango\Documents\Workspace\Data_Science\Getting_Knowing_Data\california_cities.csv')
cities.head()
cities["area_total_km2"].sort_values(ascending=False)
```

```python
# Extract latd: v  & longd: kinh 

lat, lon = cities['latd'], cities['longd']
population, area = cities['population_total'], cities['area_total_km2']

plt.style.use('seaborn-v0_8')
plt.figure(figsize=(8,6))
# Plot using Pyplot API
plt.scatter(lon, lat,
            c=np.log10(population), cmap='viridis',
            s=area, linewidths=0, alpha=0.7)
plt.axis('equal')
plt.xlabel('longlatitude')
plt.ylabel('latitude');
plt.colorbar(label='log$_{10}$(population)');
plt.clim(3,7);

#Create a legend for cities's sizes
area_range = [50, 100, 300, 500]

for area in area_range:
    plt.scatter([], [], s=area, alpha=0.6,
                label=str(area) + 'km$^2$')

plt.legend(labelspacing=1, title='City Area')

plt.title('California Cities: Population and Area Distribution');
```
